```python
anthropic_df = pd.read_csv("anthropic.csv")
anthropic_df = anthropic_df.drop("agent", axis=1)
anthropic_df.head(5)
```

```python
anthropic_df = anthropic_df.reset_index(drop=True)
anthropic_df["Row_ID"] = anthropic_df.index
anthropic_df.head(5)
```

```python
cols = ["Row_ID"] + [col for col in anthropic_df.columns if col != "Row_ID"]
anthropic_df = anthropic_df[cols]
anthropic_df
```

```python
anthropic_df["result_flag"] = anthropic_df["evaluation_result"].map({
    "success": 1,
    "failure": 0
})
```

```python
category = anthropic_df["category"].unique().tolist()
difficulty = anthropic_df["difficulty"].unique().tolist()
site = anthropic_df["site"].unique().tolist()
result = anthropic_df["result_flag"].unique().tolist()
print(len(site))
print(category, difficulty, result)
```

# Site-Level Statistics

```python
site_stats = (
    anthropic_df.groupby("site")["evaluation_result"]
    .value_counts()
    .unstack(fill_value=0)
    .reset_index()
)

# Total tasks and success rate per site
site_stats["total_tasks"] = site_stats.sum(axis=1, numeric_only=True)
site_stats["success_rate"] = (site_stats.get("success", 0) / site_stats["total_tasks"] * 100).round(2)

# Optional: sort by success rate
site_stats = site_stats.sort_values(by="success_rate", ascending=False)

# Preview
site_stats
```

# Category-level and difficulty-level stats

```python
# Category-level success rate
category_stats = anthropic_df.groupby("category")["result_flag"].agg(["count", "sum"])
category_stats["success_rate"] = (category_stats["sum"] / category_stats["count"] * 100).round(2)
category_stats = category_stats.rename(columns={"count": "total_tasks", "sum": "num_successes"})
category_stats.reset_index(inplace=True)

# Difficulty-level success rate
difficulty_stats = anthropic_df.groupby("difficulty")["result_flag"].agg(["count", "sum"])
difficulty_stats["success_rate"] = (difficulty_stats["sum"] / difficulty_stats["count"] * 100).round(2)
difficulty_stats = difficulty_stats.rename(columns={"count": "total_tasks", "sum": "num_successes"})
difficulty_stats.reset_index(inplace=True)

# Preview
# category_stats.head()
difficulty_stats.head()
```

# Failed Tasks

```python
# Only consider failed tasks
fail_reasons = (
    anthropic_df[anthropic_df["result_flag"] == 0]["evaluation_reason"]
    .value_counts()
    .reset_index()
)

fail_reasons.columns = ["reason", "count"]

# Preview
fail_reasons
```

# Saving JSONs

```python
import os

# Create output directory if not exists
os.makedirs("data", exist_ok=True)

# Save raw task list
anthropic_df.to_json("data/tasks.json", orient="records", indent=2)

# Save aggregated stats
site_stats.to_json("data/site_stats.json", orient="records", indent=2)
category_stats.to_json("data/category_stats.json", orient="records", indent=2)
difficulty_stats.to_json("data/difficulty_stats.json", orient="records", indent=2)
fail_reasons.to_json("data/failure_reasons.json", orient="records", indent=2)
```

# Site Category Matrix

```python
site_category_matrix = (
    anthropic_df.groupby(["site", "category"])["result_flag"]
    .agg(["count", "sum"])
    .reset_index()
)

site_category_matrix["success_rate"] = (
    site_category_matrix["sum"] / site_category_matrix["count"] * 100
).round(2)

site_category_matrix = site_category_matrix.rename(columns={
    "count": "total_tasks",
    "sum": "num_successes"
})

# Optional: sort by success rate
site_category_matrix = site_category_matrix.sort_values(by="success_rate", ascending=False)

# Save
site_category_matrix.to_json("data/site_category_matrix.json", orient="records", indent=2)
```

# Category Difficulty Matrix 

```python
cat_diff_matrix = (
    anthropic_df.groupby(["category", "difficulty"])["result_flag"]
    .agg(["count", "sum"])
    .reset_index()
)

cat_diff_matrix["success_rate"] = (
    cat_diff_matrix["sum"] / cat_diff_matrix["count"] * 100
).round(2)

cat_diff_matrix.rename(columns={
    "count": "total_tasks",
    "sum": "num_successes"
}, inplace=True)

cat_diff_matrix.to_json("data/category_difficulty_matrix.json", orient="records", indent=2)
```

# Most Failed Categories & Difficulties

```python
# Most failed categories (by failure rate)
category_failure_stats = (
    anthropic_df.groupby("category")["result_flag"]
    .agg(["count", "sum"])
    .reset_index()
)

category_failure_stats["failures"] = category_failure_stats["count"] - category_failure_stats["sum"]
category_failure_stats["failure_rate"] = (
    category_failure_stats["failures"] / category_failure_stats["count"] * 100
).round(2)

most_failed_categories = category_failure_stats.sort_values(by="failure_rate", ascending=False)

most_failed_categories.to_json("data/most_failed_categories.json", orient="records", indent=2)

# Most failed difficulties (by failure rate)
difficulty_failure_stats = (
    anthropic_df.groupby("difficulty")["result_flag"]
    .agg(["count", "sum"])
    .reset_index()
)

difficulty_failure_stats["failures"] = difficulty_failure_stats["count"] - difficulty_failure_stats["sum"]
difficulty_failure_stats["failure_rate"] = (
    difficulty_failure_stats["failures"] / difficulty_failure_stats["count"] * 100
).round(2)

most_failed_difficulties = difficulty_failure_stats.sort_values(by="failure_rate", ascending=False)

most_failed_difficulties.to_json("data/most_failed_difficulties.json", orient="records", indent=2)
```

# Top Failure Reasons by Category

```python
fail_patterns = (
    anthropic_df[anthropic_df["result_flag"] == 0]
    .groupby(["category", "evaluation_reason"])
    .size()
    .reset_index(name="count")
    .sort_values(by=["category", "count"], ascending=[True, False])
)

top_failures_by_category = (
    fail_patterns.groupby("category")
    .head(5)
    .reset_index(drop=True)
)

top_failures_by_category.to_json("data/top_failures_by_category.json", orient="records", indent=2)
```

```python
def compute_meta_lists(df):
    return {
        "sites": sorted(df["site"].dropna().unique().tolist()),
        "categories": sorted(df["category"].dropna().unique().tolist()),
        "difficulties": sorted(df["difficulty"].dropna().unique().tolist()),
        "evaluation_result": sorted(df["evaluation_result"].dropna().unique().tolist())
    }
```
